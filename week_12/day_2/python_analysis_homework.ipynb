#### 2. Summary stats for `Calories`

```python
drinks_content['Calories'].describe()
```

#### 3. Make new `drinks` dataframe

```python
drinks = drinks_content.loc[:,['Beverage_category', 'Beverage', 'Beverage_prep', 'Calories']].copy()
```

```python
drinks.isna().sum()
```

#### 4. create `calorie_diff` variable, 135 minus `calories`

```python
drinks.loc[:,"calorie_diff"] = 135 - drinks["Calories"]
```

#### 5. Summarising mean calories in each beverage category

```python
drinks.groupby("Beverage_category")["Calories"].mean()
```

#### 6.  Which `Beverage_prep` type contains more than the average calories of all drinks

```python
avg_cal = drinks.Calories.mean()
bev_prep_avg_cal = drinks.groupby("Beverage_prep", as_index=False)["Calories"].mean()
```

```python
bev_prep_avg_cal.loc[bev_prep_avg_cal["Calories"] > avg_cal]
```

#### 7. Which is the best type of _coffee drink_, when worried about calores

```python
drinks.Beverage_category.unique()
```

```python
drinks.loc[drinks["Beverage_category"].isin(["Coffee", "Espresso"])]
```

```python
drinks[drinks['Beverage_category'].str.contains("Coffee")].sort_values("Calories").head(1)
```

#### 8. Create new dataframe called `espresso_drinks`, with only "Classic Espresso Drinks" included

```python
espresso_drinks = drinks.loc[drinks["Beverage_category"] == "Classic Espresso Drinks"].copy()
```

#### 9. Grouping by `Beverage_prep` and find mean cal for each drink group

```python
espresso_drinks.groupby("Beverage_prep").Calories.mean()
```

#### 10. Show above results in descending order

```python
espresso_drinks.groupby("Beverage_prep", as_index=False).Calories.mean().sort_values("Calories", ascending = False)
```
